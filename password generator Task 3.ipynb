```python
import random
import string
print("welcome to your Random password generator")
length = int(input("enter the length of the password: "))
lower_d = string.ascii_lowercase
upper_d = string.ascii_uppercase
num_d = string.digits
symbols = string.punctuation
all = lower_d + upper_d + num_d + symbols
x = random.sample(all, length)
password = "".join(x)
print(password)
```
